```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
P = 0.5

np.random.seed(0)
coin = np.random.binomial(n=1, p=P, size=100)
phat = np.sum(coin) / coin.shape[0]

px = np.linspace(0.00001, 0.9999, 1000)
py = (np.sum(coin) * np.log(px)) + ((coin.shape[0] - np.sum(coin)) * np.log(1 - px))
py = np.exp(py)
plt.plot(px, py)

print(px[np.argmax(py)])
print(phat)
print(P)
```
